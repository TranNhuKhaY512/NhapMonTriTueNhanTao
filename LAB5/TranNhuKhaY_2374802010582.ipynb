#### Padding v Stride
- **Padding ($P$)**: Thm s 0 quanh nh  gi kch thc u ra. Nu $P=0$ (nh v d trn), feature map nh hn nh gc.
- **Stride ($S$)**: Bc trt ca b lc. Nu $S=1$ (nh v d), b lc trt tng pixel mt; nu $S=2$, n nhy cch 2 pixel.

Cng thc tnh kch thc u ra:

$$ \text{Output Height} = \frac{H - F + 2P}{S} + 1 $$

$$ \text{Output Width} = \frac{W - F + 2P}{S} + 1 $$

- $H$, $W$: Chiu cao v chiu rng ca nh gc.
- V d: Vi $H=W=6$, $F=3$, $P=0$, $S=1$, ta c $\frac{6-3+0}{1}+1 = 4$, ng vi kch thc feature map $4 \times 4$.

- Ly gi tr ln nht trong mt vng nh, thng l $2 \times 2$.

#### Cng thc
$$ P(i, j) = \max(I(2i:2i+2, 2j:2j+2)) $$

**Gii thch n gin**: Chia feature map thnh cc  $2 \times 2$, chn s ln nht trong mi   to feature map nh hn.

#### V d vi s "0"
Ly feature map sau ReLU:


$$ \text{ReLU}(S) = \begin{bmatrix} 2 & 0 & 3 & 0 \\ 0 & 0 & 2 & 0 \\ 2 & 0 & 2 & 0 \\ 1 & 0 & 3 & 0 \end{bmatrix} $$

p dng Max Pooling vi vng $2 \times 2$, stride = 2 (cc vng khng chng lp):
- Vng 1 (gc trn tri): $\max(2, 0, 0, 0) = 2$
- Vng 2 (gc trn phi): $\max(3, 0, 2, 0) = 3$
- Vng 3 (gc di tri): $\max(2, 0, 1, 0) = 2$
- Vng 4 (gc di phi): $\max(2, 0, 3, 0) = 3$

Kt qu feature map sau pooling:

$$ \begin{bmatrix} 2 & 3 \\ 2 & 3 \end{bmatrix} $$

** ngha**:
- Kch thc gim t $4 \times 4$ xung $2 \times 2$.
- Gi li cc gi tr ln (nt chnh), b bt chi tit nh.

### 2.4. Tng Fully Connected (FC Layer)

y l bc cui cng, ni CNN ghp tt c c trng li  on xem nh l g.

#### Cng thc
$$ y = Wx + b $$

- $x$: Vector t feature map dui ra.
- $W$: Ma trn trng s.
- $b$: Bias ( lch).

**Gii thch n gin**: Ly feature map cui, "dui" thnh mt hng s, ri nhn vi trng s  ra kt qu phn loi.

#### V d
Feature map sau pooling:
$$ \begin{bmatrix} 2 & 3 \\ 2 & 3 \end{bmatrix} $$

Dui thnh vector $x = [2, 3, 2, 3]$ (4 phn t).

Gi s ta phn loi 2 lp (s "0" hoc khng phi "0"), vi:
- Ma trn trng s:

$$ W = \begin{bmatrix} 0.5 & 0.2 & 0.3 & 0.4 \\ -0.1 & 0.3 & 0.2 & -0.5 \end{bmatrix} $$

- Bias:

$$ b = \begin{bmatrix} 0.1 \\ 0.2 \end{bmatrix} $$

Tnh $y$:
- $y_0 = (0.5 \cdot 2 + 0.2 \cdot 3 + 0.3 \cdot 2 + 0.4 \cdot 3) + 0.1 = 1 + 0.6 + 0.6 + 1.2 + 0.1 = 3.5$
- $y_1 = (-0.1 \cdot 2 + 0.3 \cdot 3 + 0.2 \cdot 2 + -0.5 \cdot 3) + 0.2 = (-0.2) + 0.9 + 0.4 - 1.5 + 0.2 = -0.2$

Kt qu: $y = [0.8, -0.1]$.

**Bc cui - Softmax**: Chuyn $y$ thnh xc sut:
- $\text{Softmax}(y_0) = \frac{e^{3.5}}{e^{3.5} + e^{-0.2}} \approx 0.976$ (xc sut l "0").
- $\text{Softmax}(y_1) = \frac{e^{-0.2}}{e^{3.5} + e^{-0.2}} \approx 0.024$ (xc sut "khng phi 0").

### Ti d liu MNIST

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# 1. Ti d liu MNIST
transform = transforms.ToTensor()  # Chuyn nh thnh tensor
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=False)
```

### XY DNG M HNH CNN C BN

```python
# 2. nh ngha m hnh CNN
class MNIST_CNN(nn.Module):  # To lp MNIST_CNN k tha t nn.Module (lp c bn ca PyTorch)
    def __init__(self):  # Hm khi to m hnh
        super(MNIST_CNN, self).__init__()  # Gi hm khi to ca lp cha (nn.Module)
        # Tng tch chp 1: 1 knh u vo (nh en trng) -> 16 knh u ra, filter 3x3
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=0)  # Tch chp gim kch thc t 28x28 -> 26x26
        # Tng tch chp 2: 16 knh u vo -> 32 knh u ra, filter 3x3
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=0)  # Tch chp gim t 12x12 -> 10x10 (sau pooling)
        #Tng tch chp 3: 32 knh u vo t `conv2`, 64 knh u ra
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=0) 
        # Tng pooling: vng 2x2, gim kch thc mt na
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)  # V d: 26x26 -> 13x13, 10x10 -> 5x5
        # Tng fully connected: t 64 knh * 5x5 (800) -> 10 lp (0-9)
        self.fc1 = nn.Linear(64 * 1 * 1, 10) # Dui feature map thnh vector ri phn loi

    def forward(self, x):  # Hm forward nh ngha lung d liu qua cc tng
        x = self.pool(torch.relu(self.conv1(x)))  # Conv1 -> ReLU (loi gi tr m) -> Pool (gim kch thc)
        x = self.pool(torch.relu(self.conv2(x)))  # Conv2 -> ReLU -> Pool, cui cng ra 32x5x5
        x = self.pool(torch.relu(self.conv3(x)))
        x.view(-1, 64 * 1 * 1)  # Dui tensor thnh vector, -1 t ng tnh batch size
        x = x.view(x.size(0), -1) 
        x = self.fc1(x)  # Qua tng fully connected, ra 10 gi tr (logits cho 0-9)
        return x  # Tr v kt qu d on
```

### Khi to m hnh

```python
# Khi to m hnh
model = MNIST_CNN()  # To mt instance ca lp MNIST_CNN
criterion = nn.CrossEntropyLoss()  # nh ngha hm mt mt CrossEntropy (dng cho phn loi nhiu lp)
optimizer = optim.SGD(model.parameters(), lr=0.1, momentum=0.9)  # Dng SGD vi learning rate 0.01 v momentum 0.9  ti u
```

### Hun luyn m hnh

```python
# 3. Hun luyn m hnh
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Chn GPU nu c, khng th dng CPU
model.to(device)  # Chuyn m hnh sang thit b  chn (GPU/CPU)

# To hai danh sch  lu gi tr mt mt (loss) v  chnh xc (accuracy) qua tng epoch
loss_values = []  # Lu mt mt trung bnh ca mi epoch  theo di qu trnh hun luyn
accuracy_values = []  # Lu  chnh xc trung bnh ca mi epoch  nh gi hiu qu m hnh

# Hun luyn m hnh qua 5 epoch (mi epoch l mt ln duyt qua ton b d liu hun luyn)
for epoch in range(10):  # Lp 10 ln, mi ln x l ton b tp d liu
    # Khi to bin  theo di tng mt mt trong epoch ny
    running_loss = 0.0  # Bin ny cng dn mt mt ca cc batch  tnh trung bnh sau
    # Khi to bin  tnh  chnh xc
    correct = 0  # m s d on ng trong epoch
    total = 0  # m tng s mu c d on trong epoch
    
    # Lp qua tng batch (nhm 64 nh) trong tp hun luyn (train_loader)
    for images, labels in train_loader:  # images: tensor cha 64 nh, labels: tensor cha 64 nhn (0-9)
        # Chuyn nh v nhn sang thit b (GPU/CPU)  ng b vi m hnh
        images, labels = images.to(device), labels.to(device)  # m bo d liu v m hnh cng thit b
        
        # Xa gradient c t bc trc  trnh tch ly
        optimizer.zero_grad()  # Gradient phi c t v 0 trc khi tnh gradient mi
        # a nh qua m hnh  nhn d on (outputs l tensor cha logits cho 10 lp)
        outputs = model(images)  # M hnh x l nh v tr v d on
        # Tnh mt mt (loss) bng cch so snh d on (outputs) vi nhn tht (labels)
        loss = criterion(outputs, labels)  # Hm CrossEntropyLoss o lng sai lch gia d on v thc t
        # Tnh gradient ngc (backpropagation)  tm hng ti u ha trng s
        loss.backward()  # Gradient c tnh cho tt c tham s ca m hnh
        # Cp nht trng s ca m hnh da trn gradient, s dng thut ton SGD
        optimizer.step()  # iu chnh trng s  gim mt mt  bc tip theo
        
        # Cng dn gi tr mt mt ca batch ny (loss.item() tr v gi tr s ca loss)
        running_loss += loss.item()  # Tng hp mt mt  tnh trung bnh sau
        
        # Tnh  chnh xc cho batch ny
        # Ly nhn d on bng cch chn lp c gi tr logits cao nht
        _, predicted = torch.max(outputs.data, 1)  # predicted cha nhn d on (0-9) cho 64 nh
        # Cng dn s mu trong batch (thng l 64, tr batch cui c th t hn)
        total += labels.size(0)  # Tng s nh  x l
        # m s d on ng bng cch so snh predicted vi labels
        correct += (predicted == labels).sum().item()  # S d on ng trong batch
    
    # Tnh mt mt trung bnh cho epoch bng cch chia tng mt mt cho s batch
    epoch_loss = running_loss / len(train_loader)  # Trung bnh mt mt ca tt c batch
    # Tnh  chnh xc trung bnh (phn trm) cho epoch
    epoch_accuracy = 100 * correct / total  # T l d on ng trn tng s mu
    # Lu gi tr mt mt v  chnh xc  v biu  sau ny
    loss_values.append(epoch_loss)  # Thm mt mt trung bnh vo danh sch
    accuracy_values.append(epoch_accuracy)  # Thm  chnh xc vo danh sch
    # In kt qu sau mi epoch  theo di tin trnh hun luyn
    print(f"Epoch {epoch+1}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%")  # Hin th mt mt v  chnh xc
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# 1. Ti d liu MNIST
transform = transforms.ToTensor()  # Chuyn nh thnh tensor
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=False)
### XY DNG M HNH CNN C BN
# 2. nh ngha m hnh CNN
class MNIST_CNN(nn.Module):  # To lp MNIST_CNN k tha t nn.Module (lp c bn ca PyTorch)
    def __init__(self):  # Hm khi to m hnh
        super(MNIST_CNN, self).__init__()  # Gi hm khi to ca lp cha (nn.Module)
        # Tng tch chp 1: 1 knh u vo (nh en trng) -> 16 knh u ra, filter 3x3
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=0)  # Tch chp gim kch thc t 28x28 -> 26x26
        # Tng tch chp 2: 16 knh u vo -> 32 knh u ra, filter 3x3
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=0)  # Tch chp gim t 12x12 -> 10x10 (sau pooling)
        # Tng pooling: vng 2x2, gim kch thc mt na
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)  # V d: 26x26 -> 13x13, 10x10 -> 5x5
        # Tng fully connected: t 32 knh * 5x5 (800) -> 10 lp (0-9)
        self.fc1 = nn.Linear(32 * 5 * 5, 10)  # Dui feature map thnh vector ri phn loi

    def forward(self, x):  # Hm forward nh ngha lung d liu qua cc tng
        x = self.pool(torch.relu(self.conv1(x)))  # Conv1 -> ReLU (loi gi tr m) -> Pool (gim kch thc)
        x = self.pool(torch.relu(self.conv2(x)))  # Conv2 -> ReLU -> Pool, cui cng ra 32x5x5
        x = x.view(-1, 32 * 5 * 5)  # Dui tensor thnh vector, -1 t ng tnh batch size
        x = self.fc1(x)  # Qua tng fully connected, ra 10 gi tr (logits cho 0-9)
        return x  # Tr v kt qu d on
### Khi to m hnh
# Khi to m hnh
model = MNIST_CNN()  # To mt instance ca lp 
criterion = nn.CrossEntropyLoss()  # nh ngha hm mt mt CrossEntropy (dng cho phn loi nhiu lp)
optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)  # Dng SGD vi learning rate 0.01 v momentum 0.9  ti u
### Hun luyn m hnh
# 3. Hun luyn m hnh
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Chn GPU nu c, khng th dng CPU
model.to(device)  # Chuyn m hnh sang thit b  chn (GPU/CPU)

# To hai danh sch  lu gi tr mt mt (loss) v  chnh xc (accuracy) qua tng epoch
loss_values = []  # Lu mt mt trung bnh ca mi epoch  theo di qu trnh hun luyn
accuracy_values = []  # Lu  chnh xc trung bnh ca mi epoch  nh gi hiu qu m hnh

# Hun luyn m hnh qua 5 epoch (mi epoch l mt ln duyt qua ton b d liu hun luyn)
for epoch in range(10):  # Lp 5 ln, mi ln x l ton b tp d liu
    # Khi to bin  theo di tng mt mt trong epoch ny
    running_loss = 0.0  # Bin ny cng dn mt mt ca cc batch  tnh trung bnh sau
    # Khi to bin  tnh  chnh xc
    correct = 0  # m s d on ng trong epoch
    total = 0  # m tng s mu c d on trong epoch
    
    # Lp qua tng batch (nhm 64 nh) trong tp hun luyn (train_loader)
    for images, labels in train_loader:  # images: tensor cha 64 nh, labels: tensor cha 64 nhn (0-9)
        # Chuyn nh v nhn sang thit b (GPU/CPU)  ng b vi m hnh
        images, labels = images.to(device), labels.to(device)  # m bo d liu v m hnh cng thit b
        
        # Xa gradient c t bc trc  trnh tch ly
        optimizer.zero_grad()  # Gradient phi c t v 0 trc khi tnh gradient mi
        # a nh qua m hnh  nhn d on (outputs l tensor cha logits cho 10 lp)
        outputs = model(images)  # M hnh x l nh v tr v d on
        # Tnh mt mt (loss) bng cch so snh d on (outputs) vi nhn tht (labels)
        loss = criterion(outputs, labels)  # Hm CrossEntropyLoss o lng sai lch gia d on v thc t
        # Tnh gradient ngc (backpropagation)  tm hng ti u ha trng s
        loss.backward()  # Gradient c tnh cho tt c tham s ca m hnh
        # Cp nht trng s ca m hnh da trn gradient, s dng thut ton SGD
        optimizer.step()  # iu chnh trng s  gim mt mt  bc tip theo
        
        # Cng dn gi tr mt mt ca batch ny (loss.item() tr v gi tr s ca loss)
        running_loss += loss.item()  # Tng hp mt mt  tnh trung bnh sau
        
        # Tnh  chnh xc cho batch ny
        # Ly nhn d on bng cch chn lp c gi tr logits cao nht
        _, predicted = torch.max(outputs.data, 1)  # predicted cha nhn d on (0-9) cho 64 nh
        # Cng dn s mu trong batch (thng l 64, tr batch cui c th t hn)
        total += labels.size(0)  # Tng s nh  x l
        # m s d on ng bng cch so snh predicted vi labels
        correct += (predicted == labels).sum().item()  # S d on ng trong batch
    
    # Tnh mt mt trung bnh cho epoch bng cch chia tng mt mt cho s batch
    epoch_loss = running_loss / len(train_loader)  # Trung bnh mt mt ca tt c batch
    # Tnh  chnh xc trung bnh (phn trm) cho epoch
    epoch_accuracy = 100 * correct / total  # T l d on ng trn tng s mu
    # Lu gi tr mt mt v  chnh xc  v biu  sau ny
    loss_values.append(epoch_loss)  # Thm mt mt trung bnh vo danh sch
    accuracy_values.append(epoch_accuracy)  # Thm  chnh xc vo danh sch
    # In kt qu sau mi epoch  theo di tin trnh hun luyn
    print(f"Epoch {epoch+1}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%")  # Hin th mt mt v  chnh xc
plt.figure(figsize=(12, 5))

# Biu  Loss
plt.subplot(1, 2, 1)
plt.plot(range(1, 11), loss_values, marker='o', color='b', label='Training Loss')
plt.title('Training Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Biu  Accuracy
plt.subplot(1, 2, 2)
plt.plot(range(1, 11), accuracy_values, marker='o', color='g', label='Training Accuracy')
plt.title('Training Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
### nh gi m hnh
correct = 0  # m s d on ng
total = 0  # m tng s mu
with torch.no_grad():  # Tt tnh gradient v ch nh gi, khng hun luyn
    for images, labels in test_loader:  # Lp qua tng batch trong tp kim tra
        images, labels = images.to(device), labels.to(device)  # Chuyn d liu sang GPU/CPU
        outputs = model(images)  # D on trn batch nh
        _, predicted = torch.max(outputs.data, 1)  # Ly ch s lp c gi tr ln nht (d on)
        total += labels.size(0)  # Cng s mu trong batch vo tng
        correct += (predicted == labels).sum().item()  # Cng s d on ng
print(f" chnh xc trn tp test: {100 * correct / total:.2f}%")  # Tnh v in  chnh xc (%)
# Trc quan kt qu d on
def visualize_prediction():  # Hm  v 5 nh v d on
    model.eval()  # Chuyn m hnh sang ch  nh gi (tt dropout, batch norm nu c)
    images, labels = next(iter(test_loader))  # Ly mt batch t tp kim tra
    images, labels = images.to(device), labels.to(device)  # Chuyn sang GPU/CPU
    outputs = model(images)  # D on trn batch
    _, predicted = torch.max(outputs, 1)  # Ly ch s lp d on
    
    # V 5 nh u tin
    fig, axes = plt.subplots(1, 5, figsize=(15, 3))  # To li 1x5  v 5 nh
    for i in range(5):  # Lp qua 5 nh u
        axes[i].imshow(images[i].cpu().squeeze(), cmap='gray')  # V nh (chuyn v CPU, b knh 1)
        axes[i].set_title(f"D on: {predicted[i].item()}\nTht: {labels[i].item()}")  # Ghi nhn d on v tht
        axes[i].axis('off')  # Tt trc ta 
    plt.show()  # Hin th hnh

visualize_prediction()  # Gi hm  v
#Trc quan feature map
def visualize_feature_map():  # Hm  v nh gc v feature map t tng conv1
    model.eval()  # Chuyn sang ch  nh gi
    images, _ = next(iter(test_loader))  # Ly mt batch t tp kim tra (b nhn)
    img = images[0].unsqueeze(0).to(device)  # Ly nh u tin, thm chiu batch (1x1x28x28)
    
    # Tnh feature map t conv1
    conv1_output = torch.relu(model.conv1(img))  # Qua conv1 v ReLU, ra tensor 1x16x26x26
    
    # V nh gc v feature map
    plt.figure(figsize=(12, 4))  
    plt.subplot(1, 3, 1)  
    plt.title("nh gc")  
    plt.imshow(img.cpu().squeeze(), cmap='gray') 
    plt.axis('off')  
    
    # V feature map u tin
    plt.subplot(1, 3, 2)  
    plt.title("Feature Map 1")  
    plt.imshow(conv1_output[0, 0].cpu().detach().numpy(), cmap='gray')  
    plt.axis('off')  
    
    # V feature map th hai
    plt.subplot(1, 3, 3)  
    plt.title("Feature Map 2") 
    plt.imshow(conv1_output[0, 1].cpu().detach().numpy(), cmap='gray')  
    plt.axis('off') 
    plt.show() 

visualize_feature_map()  
```

### Cu 2: Thm mt tng tch chp
- **Yu cu**: Thm mt tng tch chp th ba (`conv3`) vo m hnh `MNIST_CNN`.
- **Hng dn**:
  - Trong hm `__init__`, thm `self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=0)` (32 knh u vo t `conv2`, 64 knh u ra).
  - Trong hm `forward`, thm `x = self.pool(torch.relu(self.conv3(x)))` sau dng `x = self.pool(torch.relu(self.conv2(x)))`.
  - Kch thc sau `conv3` v pooling s l 64x1x1 (v 5x5 -> 3x3 -> 1x1 sau hai ln pooling v tch chp). Sa tng `fc1` thnh `self.fc1 = nn.Linear(64 * 1 * 1, 10)` v dng `x.view(-1, 64 * 1 * 1)` tng ng.
  - Chy li code v ghi nhn  chnh xc mi trn tp test.
- Vit ngn gn (2-3 cu) v tc dng ca vic thm tng tch chp (v d: tm c trng phc tp hn, nh hng n  chnh xc).

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# 1. Ti d liu MNIST
transform = transforms.ToTensor()  # Chuyn nh thnh tensor
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=False)
### XY DNG M HNH CNN C BN
# 2. nh ngha m hnh CNN
class MNIST_CNN(nn.Module):  # To lp MNIST_CNN k tha t nn.Module (lp c bn ca PyTorch)
    def __init__(self):  # Hm khi to m hnh
        super(MNIST_CNN, self).__init__()  # Gi hm khi to ca lp cha (nn.Module)
        # Tng tch chp 1: 1 knh u vo (nh en trng) -> 16 knh u ra, filter 3x3
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=0)  # Tch chp gim kch thc t 28x28 -> 26x26
        # Tng tch chp 2: 16 knh u vo -> 32 knh u ra, filter 3x3
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=0)  # Tch chp gim t 12x12 -> 10x10 (sau pooling)
        #Tng tch chp 3: 32 knh u vo t `conv2`, 64 knh u ra
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=0) 
        # Tng pooling: vng 2x2, gim kch thc mt na
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)  # V d: 26x26 -> 13x13, 10x10 -> 5x5
        # Tng fully connected: t 64 knh * 5x5 (800) -> 10 lp (0-9)
        self.fc1 = nn.Linear(64 * 1 * 1, 10) # Dui feature map thnh vector ri phn loi

    def forward(self, x):  # Hm forward nh ngha lung d liu qua cc tng
        x = self.pool(torch.relu(self.conv1(x)))  # Conv1 -> ReLU (loi gi tr m) -> Pool (gim kch thc)
        x = self.pool(torch.relu(self.conv2(x)))  # Conv2 -> ReLU -> Pool, cui cng ra 32x5x5
        x = self.pool(torch.relu(self.conv3(x)))
        x.view(-1, 64 * 1 * 1)  # Dui tensor thnh vector, -1 t ng tnh batch size
        x = x.view(x.size(0), -1) 
        x = self.fc1(x)  # Qua tng fully connected, ra 10 gi tr (logits cho 0-9)
        return x  # Tr v kt qu d on
### Khi to m hnh
# Khi to m hnh
model = MNIST_CNN()  # To mt instance ca lp 
criterion = nn.CrossEntropyLoss()  # nh ngha hm mt mt CrossEntropy (dng cho phn loi nhiu lp)
optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)  # Dng SGD vi learning rate 0.01 v momentum 0.9  ti u
### Hun luyn m hnh
# 3. Hun luyn m hnh
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Chn GPU nu c, khng th dng CPU
model.to(device)  # Chuyn m hnh sang thit b  chn (GPU/CPU)

# To hai danh sch  lu gi tr mt mt (loss) v  chnh xc (accuracy) qua tng epoch
loss_values = []  # Lu mt mt trung bnh ca mi epoch  theo di qu trnh hun luyn
accuracy_values = []  # Lu  chnh xc trung bnh ca mi epoch  nh gi hiu qu m hnh

# Hun luyn m hnh qua 5 epoch (mi epoch l mt ln duyt qua ton b d liu hun luyn)
for epoch in range(5):  # Lp 5 ln, mi ln x l ton b tp d liu
    # Khi to bin  theo di tng mt mt trong epoch ny
    running_loss = 0.0  # Bin ny cng dn mt mt ca cc batch  tnh trung bnh sau
    # Khi to bin  tnh  chnh xc
    correct = 0  # m s d on ng trong epoch
    total = 0  # m tng s mu c d on trong epoch
    
    # Lp qua tng batch (nhm 64 nh) trong tp hun luyn (train_loader)
    for images, labels in train_loader:  # images: tensor cha 64 nh, labels: tensor cha 64 nhn (0-9)
        # Chuyn nh v nhn sang thit b (GPU/CPU)  ng b vi m hnh
        images, labels = images.to(device), labels.to(device)  # m bo d liu v m hnh cng thit b
        
        # Xa gradient c t bc trc  trnh tch ly
        optimizer.zero_grad()  # Gradient phi c t v 0 trc khi tnh gradient mi
        # a nh qua m hnh  nhn d on (outputs l tensor cha logits cho 10 lp)
        outputs = model(images)  # M hnh x l nh v tr v d on
        # Tnh mt mt (loss) bng cch so snh d on (outputs) vi nhn tht (labels)
        loss = criterion(outputs, labels)  # Hm CrossEntropyLoss o lng sai lch gia d on v thc t
        # Tnh gradient ngc (backpropagation)  tm hng ti u ha trng s
        loss.backward()  # Gradient c tnh cho tt c tham s ca m hnh
        # Cp nht trng s ca m hnh da trn gradient, s dng thut ton SGD
        optimizer.step()  # iu chnh trng s  gim mt mt  bc tip theo
        
        # Cng dn gi tr mt mt ca batch ny (loss.item() tr v gi tr s ca loss)
        running_loss += loss.item()  # Tng hp mt mt  tnh trung bnh sau
        
        # Tnh  chnh xc cho batch ny
        # Ly nhn d on bng cch chn lp c gi tr logits cao nht
        _, predicted = torch.max(outputs.data, 1)  # predicted cha nhn d on (0-9) cho 64 nh
        # Cng dn s mu trong batch (thng l 64, tr batch cui c th t hn)
        total += labels.size(0)  # Tng s nh  x l
        # m s d on ng bng cch so snh predicted vi labels
        correct += (predicted == labels).sum().item()  # S d on ng trong batch
    
    # Tnh mt mt trung bnh cho epoch bng cch chia tng mt mt cho s batch
    epoch_loss = running_loss / len(train_loader)  # Trung bnh mt mt ca tt c batch
    # Tnh  chnh xc trung bnh (phn trm) cho epoch
    epoch_accuracy = 100 * correct / total  # T l d on ng trn tng s mu
    # Lu gi tr mt mt v  chnh xc  v biu  sau ny
    loss_values.append(epoch_loss)  # Thm mt mt trung bnh vo danh sch
    accuracy_values.append(epoch_accuracy)  # Thm  chnh xc vo danh sch
    # In kt qu sau mi epoch  theo di tin trnh hun luyn
    print(f"Epoch {epoch+1}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%")  # Hin th mt mt v  chnh xc
plt.figure(figsize=(12, 5))

# Biu  Loss
plt.subplot(1, 2, 1)
plt.plot(range(1, 6), loss_values, marker='o', color='b', label='Training Loss')
plt.title('Training Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Biu  Accuracy
plt.subplot(1, 2, 2)
plt.plot(range(1, 6), accuracy_values, marker='o', color='g', label='Training Accuracy')
plt.title('Training Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
### nh gi m hnh
correct = 0  # m s d on ng
total = 0  # m tng s mu
with torch.no_grad():  # Tt tnh gradient v ch nh gi, khng hun luyn
    for images, labels in test_loader:  # Lp qua tng batch trong tp kim tra
        images, labels = images.to(device), labels.to(device)  # Chuyn d liu sang GPU/CPU
        outputs = model(images)  # D on trn batch nh
        _, predicted = torch.max(outputs.data, 1)  # Ly ch s lp c gi tr ln nht (d on)
        total += labels.size(0)  # Cng s mu trong batch vo tng
        correct += (predicted == labels).sum().item()  # Cng s d on ng
print(f" chnh xc trn tp test: {100 * correct / total:.2f}%")  # Tnh v in  chnh xc (%)
# Trc quan kt qu d on
def visualize_prediction():  # Hm  v 5 nh v d on
    model.eval()  # Chuyn m hnh sang ch  nh gi (tt dropout, batch norm nu c)
    images, labels = next(iter(test_loader))  # Ly mt batch t tp kim tra
    images, labels = images.to(device), labels.to(device)  # Chuyn sang GPU/CPU
    outputs = model(images)  # D on trn batch
    _, predicted = torch.max(outputs, 1)  # Ly ch s lp d on
    
    # V 5 nh u tin
    fig, axes = plt.subplots(1, 5, figsize=(15, 3))  # To li 1x5  v 5 nh
    for i in range(5):  # Lp qua 5 nh u
        axes[i].imshow(images[i].cpu().squeeze(), cmap='gray')  # V nh (chuyn v CPU, b knh 1)
        axes[i].set_title(f"D on: {predicted[i].item()}\nTht: {labels[i].item()}")  # Ghi nhn d on v tht
        axes[i].axis('off')  # Tt trc ta 
    plt.show()  # Hin th hnh

visualize_prediction()  # Gi hm  v
#Trc quan feature map
def visualize_feature_map():  # Hm  v nh gc v feature map t tng conv1
    model.eval()  # Chuyn sang ch  nh gi
    images, _ = next(iter(test_loader))  # Ly mt batch t tp kim tra (b nhn)
    img = images[0].unsqueeze(0).to(device)  # Ly nh u tin, thm chiu batch (1x1x28x28)
    
    # Tnh feature map t conv1
    conv1_output = torch.relu(model.conv1(img))  # Qua conv1 v ReLU, ra tensor 1x16x26x26
    
    # V nh gc v feature map
    plt.figure(figsize=(12, 4))  
    plt.subplot(1, 3, 1)  
    plt.title("nh gc")  
    plt.imshow(img.cpu().squeeze(), cmap='gray') 
    plt.axis('off')  
    
    # V feature map u tin
    plt.subplot(1, 3, 2)  
    plt.title("Feature Map 1")  
    plt.imshow(conv1_output[0, 0].cpu().detach().numpy(), cmap='gray')  
    plt.axis('off')  
    
    # V feature map th hai
    plt.subplot(1, 3, 3)  
    plt.title("Feature Map 2") 
    plt.imshow(conv1_output[0, 1].cpu().detach().numpy(), cmap='gray')  
    plt.axis('off') 
    plt.show() 

visualize_feature_map()  
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# 1. Ti d liu MNIST
transform = transforms.ToTensor()  # Chuyn nh thnh tensor
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=False)
### XY DNG M HNH CNN C BN
# 2. nh ngha m hnh CNN
class MNIST_CNN(nn.Module):  # To lp MNIST_CNN k tha t nn.Module (lp c bn ca PyTorch)
    def __init__(self):  # Hm khi to m hnh
        super(MNIST_CNN, self).__init__()  # Gi hm khi to ca lp cha (nn.Module)
        # Tng tch chp 1: 1 knh u vo (nh en trng) -> 16 knh u ra, filter 3x3
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=0)  # Tch chp gim kch thc t 28x28 -> 26x26
        # Tng tch chp 2: 16 knh u vo -> 32 knh u ra, filter 3x3
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=0)  # Tch chp gim t 12x12 -> 10x10 (sau pooling)
        # Tng pooling: vng 2x2, gim kch thc mt na
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)  # V d: 26x26 -> 13x13, 10x10 -> 5x5
        # Tng fully connected: t 32 knh * 5x5 (800) -> 10 lp (0-9)
        self.fc1 = nn.Linear(32 * 5 * 5, 10)  # Dui feature map thnh vector ri phn loi

    def forward(self, x):  # Hm forward nh ngha lung d liu qua cc tng
        x = self.pool(torch.relu(self.conv1(x)))  # Conv1 -> ReLU (loi gi tr m) -> Pool (gim kch thc)
        x = self.pool(torch.relu(self.conv2(x)))  # Conv2 -> ReLU -> Pool, cui cng ra 32x5x5
        x = x.view(-1, 32 * 5 * 5)  # Dui tensor thnh vector, -1 t ng tnh batch size
        x = self.fc1(x)  # Qua tng fully connected, ra 10 gi tr (logits cho 0-9)
        return x  # Tr v kt qu d on
### Khi to m hnh
# Khi to m hnh
model = MNIST_CNN()  # To mt instance ca lp 
criterion = nn.CrossEntropyLoss()  # nh ngha hm mt mt CrossEntropy (dng cho phn loi nhiu lp)
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)  # Dng SGD vi learning rate 0.01 v momentum 0.9  ti u
### Hun luyn m hnh
# 3. Hun luyn m hnh
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Chn GPU nu c, khng th dng CPU
model.to(device)  # Chuyn m hnh sang thit b  chn (GPU/CPU)

# To hai danh sch  lu gi tr mt mt (loss) v  chnh xc (accuracy) qua tng epoch
loss_values = []  # Lu mt mt trung bnh ca mi epoch  theo di qu trnh hun luyn
accuracy_values = []  # Lu  chnh xc trung bnh ca mi epoch  nh gi hiu qu m hnh

# Hun luyn m hnh qua 5 epoch (mi epoch l mt ln duyt qua ton b d liu hun luyn)
for epoch in range(5):  # Lp 5 ln, mi ln x l ton b tp d liu
    # Khi to bin  theo di tng mt mt trong epoch ny
    running_loss = 0.0  # Bin ny cng dn mt mt ca cc batch  tnh trung bnh sau
    # Khi to bin  tnh  chnh xc
    correct = 0  # m s d on ng trong epoch
    total = 0  # m tng s mu c d on trong epoch
    
    # Lp qua tng batch (nhm 64 nh) trong tp hun luyn (train_loader)
    for images, labels in train_loader:  # images: tensor cha 64 nh, labels: tensor cha 64 nhn (0-9)
        # Chuyn nh v nhn sang thit b (GPU/CPU)  ng b vi m hnh
        images, labels = images.to(device), labels.to(device)  # m bo d liu v m hnh cng thit b
        
        # Xa gradient c t bc trc  trnh tch ly
        optimizer.zero_grad()  # Gradient phi c t v 0 trc khi tnh gradient mi
        # a nh qua m hnh  nhn d on (outputs l tensor cha logits cho 10 lp)
        outputs = model(images)  # M hnh x l nh v tr v d on
        # Tnh mt mt (loss) bng cch so snh d on (outputs) vi nhn tht (labels)
        loss = criterion(outputs, labels)  # Hm CrossEntropyLoss o lng sai lch gia d on v thc t
        # Tnh gradient ngc (backpropagation)  tm hng ti u ha trng s
        loss.backward()  # Gradient c tnh cho tt c tham s ca m hnh
        # Cp nht trng s ca m hnh da trn gradient, s dng thut ton SGD
        optimizer.step()  # iu chnh trng s  gim mt mt  bc tip theo
        
        # Cng dn gi tr mt mt ca batch ny (loss.item() tr v gi tr s ca loss)
        running_loss += loss.item()  # Tng hp mt mt  tnh trung bnh sau
        
        # Tnh  chnh xc cho batch ny
        # Ly nhn d on bng cch chn lp c gi tr logits cao nht
        _, predicted = torch.max(outputs.data, 1)  # predicted cha nhn d on (0-9) cho 64 nh
        # Cng dn s mu trong batch (thng l 64, tr batch cui c th t hn)
        total += labels.size(0)  # Tng s nh  x l
        # m s d on ng bng cch so snh predicted vi labels
        correct += (predicted == labels).sum().item()  # S d on ng trong batch
    
    # Tnh mt mt trung bnh cho epoch bng cch chia tng mt mt cho s batch
    epoch_loss = running_loss / len(train_loader)  # Trung bnh mt mt ca tt c batch
    # Tnh  chnh xc trung bnh (phn trm) cho epoch
    epoch_accuracy = 100 * correct / total  # T l d on ng trn tng s mu
    # Lu gi tr mt mt v  chnh xc  v biu  sau ny
    loss_values.append(epoch_loss)  # Thm mt mt trung bnh vo danh sch
    accuracy_values.append(epoch_accuracy)  # Thm  chnh xc vo danh sch
    # In kt qu sau mi epoch  theo di tin trnh hun luyn
    print(f"Epoch {epoch+1}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%")  # Hin th mt mt v  chnh xc
plt.figure(figsize=(12, 5))

# Biu  Loss
plt.subplot(1, 2, 1)
plt.plot(range(1, 6), loss_values, marker='o', color='b', label='Training Loss')
plt.title('Training Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Biu  Accuracy
plt.subplot(1, 2, 2)
plt.plot(range(1, 6), accuracy_values, marker='o', color='g', label='Training Accuracy')
plt.title('Training Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
### nh gi m hnh
correct = 0  # m s d on ng
total = 0  # m tng s mu
with torch.no_grad():  # Tt tnh gradient v ch nh gi, khng hun luyn
    for images, labels in test_loader:  # Lp qua tng batch trong tp kim tra
        images, labels = images.to(device), labels.to(device)  # Chuyn d liu sang GPU/CPU
        outputs = model(images)  # D on trn batch nh
        _, predicted = torch.max(outputs.data, 1)  # Ly ch s lp c gi tr ln nht (d on)
        total += labels.size(0)  # Cng s mu trong batch vo tng
        correct += (predicted == labels).sum().item()  # Cng s d on ng
print(f" chnh xc trn tp test: {100 * correct / total:.2f}%")  # Tnh v in  chnh xc (%)
# Trc quan kt qu d on
def visualize_prediction():  # Hm  v 5 nh v d on
    model.eval()  # Chuyn m hnh sang ch  nh gi (tt dropout, batch norm nu c)
    images, labels = next(iter(test_loader))  # Ly mt batch t tp kim tra
    images, labels = images.to(device), labels.to(device)  # Chuyn sang GPU/CPU
    outputs = model(images)  # D on trn batch
    _, predicted = torch.max(outputs, 1)  # Ly ch s lp d on
    
    # V 5 nh u tin
    fig, axes = plt.subplots(1, 5, figsize=(15, 3))  # To li 1x5  v 5 nh
    for i in range(5):  # Lp qua 5 nh u
        axes[i].imshow(images[i].cpu().squeeze(), cmap='gray')  # V nh (chuyn v CPU, b knh 1)
        axes[i].set_title(f"D on: {predicted[i].item()}\nTht: {labels[i].item()}")  # Ghi nhn d on v tht
        axes[i].axis('off')  # Tt trc ta 
    plt.show()  # Hin th hnh

visualize_prediction()  # Gi hm  v
#Trc quan feature map
def visualize_feature_map():  # Hm  v nh gc v feature map t tng conv1
    model.eval()  # Chuyn sang ch  nh gi
    images, _ = next(iter(test_loader))  # Ly mt batch t tp kim tra (b nhn)
    img = images[0].unsqueeze(0).to(device)  # Ly nh u tin, thm chiu batch (1x1x28x28)
    
    # Tnh feature map t conv1
    conv1_output = torch.relu(model.conv1(img))  # Qua conv1 v ReLU, ra tensor 1x16x26x26
    
    # V nh gc v feature map
    plt.figure(figsize=(12, 4))  
    plt.subplot(1, 3, 1)  
    plt.title("nh gc")  
    plt.imshow(img.cpu().squeeze(), cmap='gray') 
    plt.axis('off')  
    
    # V feature map u tin
    plt.subplot(1, 3, 2)  
    plt.title("Feature Map 1")  
    plt.imshow(conv1_output[0, 0].cpu().detach().numpy(), cmap='gray')  
    plt.axis('off')  
    
    # V feature map th hai
    plt.subplot(1, 3, 3)  
    plt.title("Feature Map 2") 
    plt.imshow(conv1_output[0, 1].cpu().detach().numpy(), cmap='gray')  
    plt.axis('off') 
    plt.show() 

visualize_feature_map()  
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# 1. Ti d liu MNIST
transform = transforms.ToTensor()  # Chuyn nh thnh tensor
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=False)
### XY DNG M HNH CNN C BN
# 2. nh ngha m hnh CNN
class MNIST_CNN(nn.Module):  # To lp MNIST_CNN k tha t nn.Module (lp c bn ca PyTorch)
    def __init__(self):  # Hm khi to m hnh
        super(MNIST_CNN, self).__init__()  # Gi hm khi to ca lp cha (nn.Module)
        # Tng tch chp 1: 1 knh u vo (nh en trng) -> 16 knh u ra, filter 3x3
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=0)  # Tch chp gim kch thc t 28x28 -> 26x26
        # Tng tch chp 2: 16 knh u vo -> 32 knh u ra, filter 3x3
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=0)  # Tch chp gim t 12x12 -> 10x10 (sau pooling)
        # Tng pooling: vng 2x2, gim kch thc mt na
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)  # V d: 26x26 -> 13x13, 10x10 -> 5x5
        # Tng fully connected: t 32 knh * 5x5 (800) -> 10 lp (0-9)
        self.fc1 = nn.Linear(32 * 5 * 5, 10)  # Dui feature map thnh vector ri phn loi

    def forward(self, x):  # Hm forward nh ngha lung d liu qua cc tng
        x = self.pool(torch.relu(self.conv1(x)))  # Conv1 -> ReLU (loi gi tr m) -> Pool (gim kch thc)
        x = self.pool(torch.relu(self.conv2(x)))  # Conv2 -> ReLU -> Pool, cui cng ra 32x5x5
        x = x.view(-1, 32 * 5 * 5)  # Dui tensor thnh vector, -1 t ng tnh batch size
        x = self.fc1(x)  # Qua tng fully connected, ra 10 gi tr (logits cho 0-9)
        return x  # Tr v kt qu d on
### Khi to m hnh
# Khi to m hnh
model = MNIST_CNN()  # To mt instance ca lp 
criterion = nn.CrossEntropyLoss()  # nh ngha hm mt mt CrossEntropy (dng cho phn loi nhiu lp)
optimizer = optim.SGD(model.parameters(), lr=0.1, momentum=0.9)  # Dng SGD vi learning rate 0.01 v momentum 0.9  ti u
### Hun luyn m hnh
# 3. Hun luyn m hnh
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Chn GPU nu c, khng th dng CPU
model.to(device)  # Chuyn m hnh sang thit b  chn (GPU/CPU)

# To hai danh sch  lu gi tr mt mt (loss) v  chnh xc (accuracy) qua tng epoch
loss_values = []  # Lu mt mt trung bnh ca mi epoch  theo di qu trnh hun luyn
accuracy_values = []  # Lu  chnh xc trung bnh ca mi epoch  nh gi hiu qu m hnh

# Hun luyn m hnh qua 5 epoch (mi epoch l mt ln duyt qua ton b d liu hun luyn)
for epoch in range(5):  # Lp 5 ln, mi ln x l ton b tp d liu
    # Khi to bin  theo di tng mt mt trong epoch ny
    running_loss = 0.0  # Bin ny cng dn mt mt ca cc batch  tnh trung bnh sau
    # Khi to bin  tnh  chnh xc
    correct = 0  # m s d on ng trong epoch
    total = 0  # m tng s mu c d on trong epoch
    
    # Lp qua tng batch (nhm 64 nh) trong tp hun luyn (train_loader)
    for images, labels in train_loader:  # images: tensor cha 64 nh, labels: tensor cha 64 nhn (0-9)
        # Chuyn nh v nhn sang thit b (GPU/CPU)  ng b vi m hnh
        images, labels = images.to(device), labels.to(device)  # m bo d liu v m hnh cng thit b
        
        # Xa gradient c t bc trc  trnh tch ly
        optimizer.zero_grad()  # Gradient phi c t v 0 trc khi tnh gradient mi
        # a nh qua m hnh  nhn d on (outputs l tensor cha logits cho 10 lp)
        outputs = model(images)  # M hnh x l nh v tr v d on
        # Tnh mt mt (loss) bng cch so snh d on (outputs) vi nhn tht (labels)
        loss = criterion(outputs, labels)  # Hm CrossEntropyLoss o lng sai lch gia d on v thc t
        # Tnh gradient ngc (backpropagation)  tm hng ti u ha trng s
        loss.backward()  # Gradient c tnh cho tt c tham s ca m hnh
        # Cp nht trng s ca m hnh da trn gradient, s dng thut ton SGD
        optimizer.step()  # iu chnh trng s  gim mt mt  bc tip theo
        
        # Cng dn gi tr mt mt ca batch ny (loss.item() tr v gi tr s ca loss)
        running_loss += loss.item()  # Tng hp mt mt  tnh trung bnh sau
        
        # Tnh  chnh xc cho batch ny
        # Ly nhn d on bng cch chn lp c gi tr logits cao nht
        _, predicted = torch.max(outputs.data, 1)  # predicted cha nhn d on (0-9) cho 64 nh
        # Cng dn s mu trong batch (thng l 64, tr batch cui c th t hn)
        total += labels.size(0)  # Tng s nh  x l
        # m s d on ng bng cch so snh predicted vi labels
        correct += (predicted == labels).sum().item()  # S d on ng trong batch
    
    # Tnh mt mt trung bnh cho epoch bng cch chia tng mt mt cho s batch
    epoch_loss = running_loss / len(train_loader)  # Trung bnh mt mt ca tt c batch
    # Tnh  chnh xc trung bnh (phn trm) cho epoch
    epoch_accuracy = 100 * correct / total  # T l d on ng trn tng s mu
    # Lu gi tr mt mt v  chnh xc  v biu  sau ny
    loss_values.append(epoch_loss)  # Thm mt mt trung bnh vo danh sch
    accuracy_values.append(epoch_accuracy)  # Thm  chnh xc vo danh sch
    # In kt qu sau mi epoch  theo di tin trnh hun luyn
    print(f"Epoch {epoch+1}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%")  # Hin th mt mt v  chnh xc
plt.figure(figsize=(12, 5))

# Biu  Loss
plt.subplot(1, 2, 1)
plt.plot(range(1, 6), loss_values, marker='o', color='b', label='Training Loss')
plt.title('Training Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Biu  Accuracy
plt.subplot(1, 2, 2)
plt.plot(range(1, 6), accuracy_values, marker='o', color='g', label='Training Accuracy')
plt.title('Training Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
### nh gi m hnh
correct = 0  # m s d on ng
total = 0  # m tng s mu
with torch.no_grad():  # Tt tnh gradient v ch nh gi, khng hun luyn
    for images, labels in test_loader:  # Lp qua tng batch trong tp kim tra
        images, labels = images.to(device), labels.to(device)  # Chuyn d liu sang GPU/CPU
        outputs = model(images)  # D on trn batch nh
        _, predicted = torch.max(outputs.data, 1)  # Ly ch s lp c gi tr ln nht (d on)
        total += labels.size(0)  # Cng s mu trong batch vo tng
        correct += (predicted == labels).sum().item()  # Cng s d on ng
print(f" chnh xc trn tp test: {100 * correct / total:.2f}%")  # Tnh v in  chnh xc (%)
# Trc quan kt qu d on
def visualize_prediction():  # Hm  v 5 nh v d on
    model.eval()  # Chuyn m hnh sang ch  nh gi (tt dropout, batch norm nu c)
    images, labels = next(iter(test_loader))  # Ly mt batch t tp kim tra
    images, labels = images.to(device), labels.to(device)  # Chuyn sang GPU/CPU
    outputs = model(images)  # D on trn batch
    _, predicted = torch.max(outputs, 1)  # Ly ch s lp d on
    
    # V 5 nh u tin
    fig, axes = plt.subplots(1, 5, figsize=(15, 3))  # To li 1x5  v 5 nh
    for i in range(5):  # Lp qua 5 nh u
        axes[i].imshow(images[i].cpu().squeeze(), cmap='gray')  # V nh (chuyn v CPU, b knh 1)
        axes[i].set_title(f"D on: {predicted[i].item()}\nTht: {labels[i].item()}")  # Ghi nhn d on v tht
        axes[i].axis('off')  # Tt trc ta 
    plt.show()  # Hin th hnh

visualize_prediction()  # Gi hm  v
#Trc quan feature map
def visualize_feature_map():  # Hm  v nh gc v feature map t tng conv1
    model.eval()  # Chuyn sang ch  nh gi
    images, _ = next(iter(test_loader))  # Ly mt batch t tp kim tra (b nhn)
    img = images[0].unsqueeze(0).to(device)  # Ly nh u tin, thm chiu batch (1x1x28x28)
    
    # Tnh feature map t conv1
    conv1_output = torch.relu(model.conv1(img))  # Qua conv1 v ReLU, ra tensor 1x16x26x26
    
    # V nh gc v feature map
    plt.figure(figsize=(12, 4))  
    plt.subplot(1, 3, 1)  
    plt.title("nh gc")  
    plt.imshow(img.cpu().squeeze(), cmap='gray') 
    plt.axis('off')  
    
    # V feature map u tin
    plt.subplot(1, 3, 2)  
    plt.title("Feature Map 1")  
    plt.imshow(conv1_output[0, 0].cpu().detach().numpy(), cmap='gray')  
    plt.axis('off')  
    
    # V feature map th hai
    plt.subplot(1, 3, 3)  
    plt.title("Feature Map 2") 
    plt.imshow(conv1_output[0, 1].cpu().detach().numpy(), cmap='gray')  
    plt.axis('off') 
    plt.show() 

visualize_feature_map()  
```

### Cu 4: V thm feature map t tng tch chp th hai
- **Yu cu**: Sa hm `visualize_feature_map`  v thm hai feature map t tng `conv2`.
- **Hng dn**:
  - Trong hm `visualize_feature_map`, thm dng `conv2_output = torch.relu(self.conv2(self.pool(torch.relu(self.conv1(img)))))`  tnh feature map t `conv2`.
  - Tng khung hnh t 3 ct thnh 5 ct: `plt.figure(figsize=(20, 4))` v sa cc subplot thnh `plt.subplot(1, 5, ...)`.
  - Thm hai subplot  v `conv2_output[0, 0]` v `conv2_output[0, 1]` (tng t nh `conv1_output`).
  - Chy li v m t s khc bit gia feature map t `conv1` v `conv2` (v d: chi tit hn, tru tng hn).
- Vit ngn gn (2-3 cu) v s khc bit gia feature map t cc tng khc nhau.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# 1. Ti d liu MNIST
transform = transforms.ToTensor()  # Chuyn nh thnh tensor
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=False)
### XY DNG M HNH CNN C BN
# 2. nh ngha m hnh CNN
class MNIST_CNN(nn.Module):  # To lp MNIST_CNN k tha t nn.Module (lp c bn ca PyTorch)
    def __init__(self):  # Hm khi to m hnh
        super(MNIST_CNN, self).__init__()  # Gi hm khi to ca lp cha (nn.Module)
        # Tng tch chp 1: 1 knh u vo (nh en trng) -> 16 knh u ra, filter 3x3
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=0)  # Tch chp gim kch thc t 28x28 -> 26x26
        # Tng tch chp 2: 16 knh u vo -> 32 knh u ra, filter 3x3
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=0)  # Tch chp gim t 12x12 -> 10x10 (sau pooling)
        # Tng pooling: vng 2x2, gim kch thc mt na
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)  # V d: 26x26 -> 13x13, 10x10 -> 5x5
        # Tng fully connected: t 32 knh * 5x5 (800) -> 10 lp (0-9)
        self.fc1 = nn.Linear(32 * 5 * 5, 10)  # Dui feature map thnh vector ri phn loi

    def forward(self, x):  # Hm forward nh ngha lung d liu qua cc tng
        x = self.pool(torch.relu(self.conv1(x)))  # Conv1 -> ReLU (loi gi tr m) -> Pool (gim kch thc)
        x = self.pool(torch.relu(self.conv2(x)))  # Conv2 -> ReLU -> Pool, cui cng ra 32x5x5
        x = x.view(-1, 32 * 5 * 5)  # Dui tensor thnh vector, -1 t ng tnh batch size
        x = self.fc1(x)  # Qua tng fully connected, ra 10 gi tr (logits cho 0-9)
        return x  # Tr v kt qu d on
### Khi to m hnh
# Khi to m hnh
model = MNIST_CNN()  # To mt instance ca lp 
criterion = nn.CrossEntropyLoss()  # nh ngha hm mt mt CrossEntropy (dng cho phn loi nhiu lp)
optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)  # Dng SGD vi learning rate 0.01 v momentum 0.9  ti u
### Hun luyn m hnh
# 3. Hun luyn m hnh
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Chn GPU nu c, khng th dng CPU
model.to(device)  # Chuyn m hnh sang thit b  chn (GPU/CPU)

# To hai danh sch  lu gi tr mt mt (loss) v  chnh xc (accuracy) qua tng epoch
loss_values = []  # Lu mt mt trung bnh ca mi epoch  theo di qu trnh hun luyn
accuracy_values = []  # Lu  chnh xc trung bnh ca mi epoch  nh gi hiu qu m hnh

# Hun luyn m hnh qua 5 epoch (mi epoch l mt ln duyt qua ton b d liu hun luyn)
for epoch in range(5):  # Lp 5 ln, mi ln x l ton b tp d liu
    # Khi to bin  theo di tng mt mt trong epoch ny
    running_loss = 0.0  # Bin ny cng dn mt mt ca cc batch  tnh trung bnh sau
    # Khi to bin  tnh  chnh xc
    correct = 0  # m s d on ng trong epoch
    total = 0  # m tng s mu c d on trong epoch
    
    # Lp qua tng batch (nhm 64 nh) trong tp hun luyn (train_loader)
    for images, labels in train_loader:  # images: tensor cha 64 nh, labels: tensor cha 64 nhn (0-9)
        # Chuyn nh v nhn sang thit b (GPU/CPU)  ng b vi m hnh
        images, labels = images.to(device), labels.to(device)  # m bo d liu v m hnh cng thit b
        
        # Xa gradient c t bc trc  trnh tch ly
        optimizer.zero_grad()  # Gradient phi c t v 0 trc khi tnh gradient mi
        # a nh qua m hnh  nhn d on (outputs l tensor cha logits cho 10 lp)
        outputs = model(images)  # M hnh x l nh v tr v d on
        # Tnh mt mt (loss) bng cch so snh d on (outputs) vi nhn tht (labels)
        loss = criterion(outputs, labels)  # Hm CrossEntropyLoss o lng sai lch gia d on v thc t
        # Tnh gradient ngc (backpropagation)  tm hng ti u ha trng s
        loss.backward()  # Gradient c tnh cho tt c tham s ca m hnh
        # Cp nht trng s ca m hnh da trn gradient, s dng thut ton SGD
        optimizer.step()  # iu chnh trng s  gim mt mt  bc tip theo
        
        # Cng dn gi tr mt mt ca batch ny (loss.item() tr v gi tr s ca loss)
        running_loss += loss.item()  # Tng hp mt mt  tnh trung bnh sau
        
        # Tnh  chnh xc cho batch ny
        # Ly nhn d on bng cch chn lp c gi tr logits cao nht
        _, predicted = torch.max(outputs.data, 1)  # predicted cha nhn d on (0-9) cho 64 nh
        # Cng dn s mu trong batch (thng l 64, tr batch cui c th t hn)
        total += labels.size(0)  # Tng s nh  x l
        # m s d on ng bng cch so snh predicted vi labels
        correct += (predicted == labels).sum().item()  # S d on ng trong batch
    
    # Tnh mt mt trung bnh cho epoch bng cch chia tng mt mt cho s batch
    epoch_loss = running_loss / len(train_loader)  # Trung bnh mt mt ca tt c batch
    # Tnh  chnh xc trung bnh (phn trm) cho epoch
    epoch_accuracy = 100 * correct / total  # T l d on ng trn tng s mu
    # Lu gi tr mt mt v  chnh xc  v biu  sau ny
    loss_values.append(epoch_loss)  # Thm mt mt trung bnh vo danh sch
    accuracy_values.append(epoch_accuracy)  # Thm  chnh xc vo danh sch
    # In kt qu sau mi epoch  theo di tin trnh hun luyn
    print(f"Epoch {epoch+1}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%")  # Hin th mt mt v  chnh xc
plt.figure(figsize=(12, 5))

# Biu  Loss
plt.subplot(1, 2, 1)
plt.plot(range(1, 6), loss_values, marker='o', color='b', label='Training Loss')
plt.title('Training Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Biu  Accuracy
plt.subplot(1, 2, 2)
plt.plot(range(1, 6), accuracy_values, marker='o', color='g', label='Training Accuracy')
plt.title('Training Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
### nh gi m hnh
correct = 0  # m s d on ng
total = 0  # m tng s mu
with torch.no_grad():  # Tt tnh gradient v ch nh gi, khng hun luyn
    for images, labels in test_loader:  # Lp qua tng batch trong tp kim tra
        images, labels = images.to(device), labels.to(device)  # Chuyn d liu sang GPU/CPU
        outputs = model(images)  # D on trn batch nh
        _, predicted = torch.max(outputs.data, 1)  # Ly ch s lp c gi tr ln nht (d on)
        total += labels.size(0)  # Cng s mu trong batch vo tng
        correct += (predicted == labels).sum().item()  # Cng s d on ng
print(f" chnh xc trn tp test: {100 * correct / total:.2f}%")  # Tnh v in  chnh xc (%)
# Trc quan kt qu d on
def visualize_prediction():  # Hm  v 5 nh v d on
    model.eval()  # Chuyn m hnh sang ch  nh gi (tt dropout, batch norm nu c)
    images, labels = next(iter(test_loader))  # Ly mt batch t tp kim tra
    images, labels = images.to(device), labels.to(device)  # Chuyn sang GPU/CPU
    outputs = model(images)  # D on trn batch
    _, predicted = torch.max(outputs, 1)  # Ly ch s lp d on
    
    # V 5 nh u tin
    fig, axes = plt.subplots(1, 5, figsize=(15, 3))  # To li 1x5  v 5 nh
    for i in range(5):  # Lp qua 5 nh u
        axes[i].imshow(images[i].cpu().squeeze(), cmap='gray')  # V nh (chuyn v CPU, b knh 1)
        axes[i].set_title(f"D on: {predicted[i].item()}\nTht: {labels[i].item()}")  # Ghi nhn d on v tht
        axes[i].axis('off')  # Tt trc ta 
    plt.show()  # Hin th hnh

visualize_prediction()  # Gi hm  v
#Trc quan feature map
def visualize_feature_map():  # Hm  v nh gc v feature map t tng conv1
    model.eval()  # Chuyn sang ch  nh gi
    images, _ = next(iter(test_loader))  # Ly mt batch t tp kim tra (b nhn)
    img = images[0].unsqueeze(0).to(device)  # Ly nh u tin, thm chiu batch (1x1x28x28)
    
    # Tnh feature map t conv1
    conv1_output = torch.relu(model.conv1(img))  # Qua conv1 v ReLU, ra tensor 1x16x26x26
    x = model.pool(conv1_output)                      # 1  16  13  13
    conv2_output = torch.relu(model.conv2(x))   
    # V nh gc v feature map
    plt.figure(figsize=(20, 4)) 
    plt.subplot(1, 5, 1)  
    plt.title("nh gc")  
    plt.imshow(img.cpu().squeeze(), cmap='gray') 
    plt.axis('off')  
    
    # V feature map u tin
    plt.subplot(1, 5, 2)  
    plt.title("Feature Map Conv1 - 1")  
    plt.imshow(conv1_output[0, 0].cpu().detach().numpy(), cmap='gray')  
    plt.axis('off')  
    
    # V feature map th hai
    plt.subplot(1, 5, 3)  
    plt.title("Feature Map Conv1-2") 
    plt.imshow(conv1_output[0, 1].cpu().detach().numpy(), cmap='gray')  
    plt.axis('off') 
   

     # Hai feature map u ca conv2
    plt.subplot(1, 5, 4)
    plt.imshow(conv2_output[0, 0].cpu().detach(), cmap='gray')
    plt.title("Feature MAP Conv2-1");  
    plt.axis('off') 

    plt.subplot(1, 5, 5)
    plt.imshow(conv2_output[0, 1].cpu().detach(), cmap='gray')
    plt.title("Feature MAP Conv2-2 "); 
    plt.axis('off') 

    plt.show() 
visualize_feature_map()  
```
