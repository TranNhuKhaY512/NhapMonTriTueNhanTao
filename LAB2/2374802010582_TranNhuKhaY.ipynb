## Cơ Sở Lý Thuyết

### 1. Tìm Kiếm Mù
Tìm kiếm mù (Uninformed Search) là các thuật toán tìm kiếm không sử dụng thông tin heuristic, duyệt qua không gian trạng thái theo cách có hệ thống, dựa trên cấu trúc bài toán.

### 2. Tìm Kiếm theo Chiều Rộng (BFS)
- **Nguyên lý**: Khám phá tất cả các nút ở mức hiện tại trước khi chuyển sang mức tiếp theo.
- **Cấu trúc dữ liệu**: Hàng đợi (Queue).
- **Đặc điểm**:
  - Trong đồ thị không trọng số, BFS đảm bảo tìm đường đi ngắn nhất (theo số cạnh).
  - Trong đồ thị có trọng số, BFS không đảm bảo đường đi có tổng trọng số nhỏ nhất, vì nó không xem xét trọng số của các cạnh.

### 3. Tìm Kiếm theo Chiều Sâu (DFS)
- **Nguyên lý**: Khám phá một nhánh đến tận cùng trước khi quay lại thử nhánh khác.
- **Cấu trúc dữ liệu**: Ngăn xếp (Stack) hoặc đệ quy.
- **Đặc điểm**:
  - DFS không đảm bảo đường đi ngắn nhất trong cả đồ thị không trọng số và có trọng số.
  - Kết quả phụ thuộc vào thứ tự duyệt các nút kề.

### 4. Đồ Thị Không Trọng Số và Có Trọng Số
- **Đồ thị không trọng số**: Các cạnh có trọng số ngầm định bằng 1. BFS tối ưu theo số cạnh.
- **Đồ thị có trọng số**: Các cạnh có trọng số khác nhau (ví dụ: chi phí, khoảng cách). BFS và DFS không được thiết kế để tối ưu hóa tổng trọng số, dẫn đến kết quả không tối ưu.

### 5. Bài Toán 
Cho các đồ thị vô hướng (không trọng số và có trọng số) biểu diễn mạng lưới hoặc mê cung. Nhiệm vụ là tìm đường đi từ nút bắt đầu (**S**) đến nút đích (**G** hoặc **H**) bằng BFS và DFS, thông qua minh họa thủ công và lập trình với chú thích chi tiết.

2. Lấy S, thêm A, B, C: Hàng đợi = [(A, [S, A]), (B, [S, B]), (C, [S, C])], Đã thăm = {S, A, B, C}

**Minh họa DFS** :
1. Khởi tạo: Ngăn xếp = [(S, [S])], Đã thăm = {S}

3. Lấy A, thêm D: Ngăn xếp = [(D, [S, A, D])], Đã thăm = {S, A, D}
4. Lấy D, thêm F: Ngăn xếp = [(F, [S, A, D, F])], Đã thăm = {S, A, D, F}
5. Lấy F, không có kề mới: Ngăn xếp rỗng, quay lại S.
6. Từ S, thêm B: Ngăn xếp = [(B, [S, B])], Đã thăm = {S, A, D, F, B}
7. Lấy B, thêm E: Ngăn xếp = [(E, [S, B, E])], Đã thăm = {S, A, D, F, B, E}
8. Lấy E, thêm G: G là đích, trả về [S, B, E, G]

**Kết quả**: Đường đi DFS: S → B → E → G.

### Phần 3: Triển Khai Mã Python

#### BFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 1**.

````python
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']

 
}

# Chạy BFS và in kết quả
path = bfs(graph1, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 1:", path)
````

#### DFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 1**.

````python
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy DFS và in kết quả
path = dfs(graph1, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 1:", path)
````

#### BFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 5**.

````python
from collections import deque

def bfs_weighted(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu, đường đi, và tổng trọng số
    queue = deque([(start, [start], 0)])
    # Khởi tạo tập hợp các nút đã thăm
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút, đường đi, và tổng trọng số từ đầu hàng đợi
        (node, path, total_weight) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path, total_weight
        # Duyệt qua các nút kề và trọng số tương ứng
        for neighbor, weight in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề, đường đi mới, và tổng trọng số cập nhật vào hàng đợi
                queue.append((neighbor, path + [neighbor], total_weight + weight))
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy BFS và in kết quả
path, weight = bfs_weighted(graph5, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

#### DFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 5**.

````python
def dfs_weighted(graph, start, goal, visited=None, path=None, total_weight=0):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path, total_weight
    
    # Duyệt qua các nút kề và trọng số tương ứng
    for neighbor, weight in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi và tổng trọng số mới
            new_path, new_weight = dfs_weighted(graph, neighbor, goal, visited, path + [neighbor], total_weight + weight)
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path, new_weight
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy DFS và in kết quả
path, weight = dfs_weighted(graph5, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

### Trả lời câu hỏi lý thuyết ở đây:

#### 1. Với đồ thì không có trọng số: 
- Mọi cạnh cuả đồ thị được xem là có trọng số bằng nhau.
- BFS duyệt theo từng lớp (các đỉnh cách nhau bắt đầu từ 1 cạnh rồi đến 2 cạnh ,...)
- Do đó BFS luôn tìm được đường đi có ít số cạnh nhất (đường đi ngắn nhất về số cạnh) từ đỉnh bắt đầu đến đích
##### Với đồ thị có trọng số thì
- BFS không xét trọng số cạnh khi duyệt.
- Tuy vậy nó vẫn chọn đường đi theo số cạnh ít nhất và bỏ qua tổng trọng số, nên đôi khi có thể chọn ra đường có trọng số dài hơn.
##### Dựa vào minh họa đồ thị mẫu 5
- Với mục tiêu tìm đường từ S → G
- Level 1: S→A(1), B(2), C(3)
- Level 2: A → G(3) (trọng số là 1+3 =4)
           B→G (trọng số là 2 +5 =7)
           S→A→D→G (1+4+6=11)
- BFS sẽ lấy G khi qua A đầu tiên và chọn đường S → A→G (chỉ cần 2 bước).
- Tuy nhiên, trong trường hợp trọng số BFS không đảm bảo tìm được đường đi có tổng trọng số ít nhất do ưu tiên đường có ít số cạnh hơn.

#### 2. Để tránh lặp vô hạn trong đồ thị có chu trình thì:
- BFS và DFS đã sử dụng "đã thăm".
- Mỗi khi đến 1 đỉnh thì nó đánh dấu đỉnh đó đã thăm.
- Và trước khi thêm đỉnh mới vào hàng đợi (BFS) hoặc ngăn xếp(DFS) thì nó sẽ kiểm tra đỉnh đó đã thăm hay chưa.
- Nếu đã thăm thì bỏ qua, tránh quay lại, tránh được lặp vô hạn.
##### VD từ đồ thị mẫu 2
- Nếu BFS hoặc DFS không dùng "đã thăm" thì sẽ bị lặp giữa A và B.

#### Bài Tập Lập Trình: BFS trên Đồ Thị Mẫu 2
Chạy BFS trên **Đồ thị mẫu 2** với chú thích chi tiết.

````python
# code here
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 2 (có chu trình, không trọng số)
graph1 = {
   'S': ['A', 'B'],
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': [ 'D'],
    'D': ['G'],
    'G': []

}

# Chạy BFS và in kết quả
path = bfs(graph1, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 2:", path)
````

#### Bài Tập Lập Trình: DFS trên Đồ Thị Mẫu 2
Chạy DFS trên **Đồ thị mẫu 2** với chú thích chi tiết.

````python
# code here
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 2 (có chu trình , không trọng số)
graph1 = {
    'S': ['A', 'B'],
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': [ 'D'],
    'D': ['G'],
    'G': []

}


# Chạy DFS và in kết quả
path = dfs(graph1, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 2:", path)
````

### 2. Sửa mã BFS để lưu và in tất cả các đường đi từ S đến H trên **Đồ thị mẫu 7** (không trọng số). Đảm bảo mã có chú thích đầy đủ.


````python
# Import deque để sử dụng hàng đợi (queue) một cách hiệu quả
from collections import deque

def find_all_paths(graph, start, goal):
    # Khởi tạo một danh sách để lưu tất cả các đường đi tìm được.
    all_paths = []
    # Khởi tạo hàng đợi, với nút bắt đầu và đường đi ban đầu
    queue = deque([[start]])
    # Bắt đầu vòng lặp cho đến khi hàng đợi rỗng.
    while queue:
        # Lấy đường đi đầu tiên ra khỏi hàng đợi để xử lý.
        path = queue.popleft()
        # Lấy nút cuối cùng trong đường đi hiện tại.
        node = path[-1]
        # Kiểm tra nếu nút cuối cùng này là nút đích.
        if node == goal:
            all_paths.append(path)
        
        # Duyệt qua các nút kề của nút hiện tại.
        for neighbor in graph.get(node, []):
            if neighbor not in path:
                # Tạo một đường đi mới bằng cách lấy đường đi cũ...
                new_path = list(path)
                # ...và thêm nút kề vào.
                new_path.append(neighbor)
                
                # Thêm đường đi mới này vào cuối hàng đợi 
                queue.append(new_path)
                
    # trả về danh sách tất cả các đường đi đã tìm thấy.
    return all_paths

# Định nghĩa Đồ thị mẫu 7 (không trọng số)

graph7 = {
    'S': ['A', 'D', 'E'],
    'A': ['S', 'B', 'D', 'E'],
    'B': ['A', 'C', 'E', 'F'],
    'C': ['B', 'F', 'G'],
    'D': ['S', 'A', 'E'], 
    'E': ['S', 'A', 'B', 'D', 'F', 'H'],
    'F': ['B', 'C', 'E', 'G', 'H'],
    'G': ['C', 'F', 'H'],
    'H': ['E', 'F', 'G']
}

# gọi hàm để tìm tất cả các đường đi từ 'S' đến 'H'
all_found_paths = find_all_paths(graph7, 'S', 'H')

print(f"Tất cả các đường đi từ 'S' đến 'H' trên Đồ thị mẫu 7 được tìm thấy: ")
if all_found_paths:
    # Sắp xếp kết quả theo độ dài đường đi 
    all_found_paths.sort(key=len)
    for i, path in enumerate(all_found_paths):
        print(f"  {i+1}. {' -> '.join(path)}")
else:
    print("Không tìm thấy đường đi nào.")
````

````python

#BFS đồ thị mẫu 6
from collections import deque
import time

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu, đường đi, và tổng trọng số
    queue = deque([(start, [start], 0)])
    # Khởi tạo tập hợp các nút đã thăm
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút, đường đi, và tổng trọng số từ đầu hàng đợi
        (node, path, total_weight) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path, total_weight
        # Duyệt qua các nút kề và trọng số tương ứng
        for neighbor, weight in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề, đường đi mới, và tổng trọng số cập nhật vào hàng đợi
                queue.append((neighbor, path + [neighbor], total_weight + weight))
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 6 (có trọng số)
graph6 = {
    'S': [('A', 2), ('C', 5)],
    'A': [('S', 2), ('B', 3), ('D', 4)],
    'B': [('A', 3), ('E', 6)],
    'C': [('S', 5), ('D', 7), ('F', 9)],
    'D': [('A',4)],
    'E': [('A', 6), ('H', 10)],
    'F': [('C', 9), ('G', 12)],
    'G': [('F', 12), ('H', 15)]
}
  
# Chạy BFS và in kết quả
start_time_bfs = time.perf_counter()
path, weight = bfs(graph6, 'S', 'H')
end_time_bfs = time.perf_counter()

print("Đường đi BFS trên Đồ thị mẫu 6 từ S đến H:", path)
print("Tổng trọng số:", weight)
print("Thời gian chạy: ", end_time_bfs - start_time_bfs)


# DFS đồ thị mẫu 6
def dfs(graph, start, goal, visited=None, path=None, total_weight=0):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path, total_weight
    
    # Duyệt qua các nút kề và trọng số tương ứng
    for neighbor, weight in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi và tổng trọng số mới
            new_path, new_weight = dfs(graph, neighbor, goal, visited, path + [neighbor], total_weight + weight)
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path, new_weight
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 6 (có trọng số)
graph6 = {
    'S': [('A', 2), ('C', 5)],
    'A': [('S', 2), ('B', 3), ('D', 4)],
    'B': [('A', 3), ('E', 6)],
    'C': [('S', 5), ('D', 7), ('F', 9)],
    'D': [('A',4)],
    'E': [('A', 6), ('H', 10)],
    'F': [('C', 9), ('G', 12)],
    'G': [('F', 12), ('H', 15)]
}

# Chạy DFS và in kết quả
start_time_dfs = time.perf_counter()
path_dfs, weight = dfs(graph6, 'S', 'H')
end_time_dfs = time.perf_counter()

print("Đường đi DFS trên Đồ thị mẫu 6 từ S đến H:", path_dfs)
print("Tổng trọng số:", weight)
print("Thời gian chạy: ", end_time_dfs - start_time_dfs)


````

##### So sánh hiệu suất
1. DFS có thời gian chạy nhanh hơn BFS trong ví dụ này do đi đúng nhánh chứa đích ngay từ đầu.
2. BFS sử dụng nhiều bộ nhớ hơn DFS vì phải lưu hàng đợi các nhánh mở rộng.
3. DFS duyệt ít đỉnh hơn BFS, giúp giảm số bước xử lý trong trường hợp này.
4. Cả hai đều tìm ra đường đi đúng, nhưng không đảm bảo trọng số nhỏ nhất.

##### Đồ thị mẫu 7


````python
#bfs
from collections import deque
import time

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph.get(node, []):
             # Nếu nút kề là đích, ta đã tìm thấy đường đi.
            if neighbor == goal:
                # Thêm đích vào đường đi và trả về ngay lập tức.
                return path + [neighbor]
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None


# Định nghĩa Đồ thị mẫu 7 (không trọng số)
graph7 = {
    'S': ['A', 'D', 'E'],
    'A': ['S', 'B', 'D', 'E'],
    'B': ['A', 'C', 'E', 'F'],
    'C': ['B', 'F', 'G'],
    'D': ['S', 'A', 'E'], 
    'E': ['S', 'A', 'B', 'D', 'F', 'H'],
    'F': ['B', 'C', 'E', 'G', 'H'],
    'G': ['C', 'F', 'H'],
    'H': ['E', 'F', 'G']
}

# Chạy BFS để tìm một đường đi từ S đến H
start_time_bfs = time.perf_counter()
path_bfs= bfs(graph7, 'S', 'H')
end_time_bfs = time.perf_counter()

print("Đường đi BFS trên Đồ thị mẫu 7 từ S đến H:", path_bfs)
print("Thời gian chạy: ", end_time_bfs - start_time_bfs)

#dfs
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 7 (không trọng số)
graph7 = {
    'S': ['A', 'D', 'E'],
    'A': ['S', 'B', 'D', 'E'],
    'B': ['A', 'C', 'E', 'F'],
    'C': ['B', 'F', 'G'],
    'D': ['S', 'A', 'E'], 
    'E': ['S', 'A', 'B', 'D', 'F', 'H'],
    'F': ['B', 'C', 'E', 'G', 'H'],
    'G': ['C', 'F', 'H'],
    'H': ['E', 'F', 'G']
}

# Chạy DFS và in kết quả
start_time_dfs = time.perf_counter()
path= dfs(graph7, 'S', 'H')
end_time_dfs = time.perf_counter()

print("Đường đi DFS trên Đồ thị mẫu 7 từ S đến H:", path)
print("Thời gian chạy: ", end_time_dfs - start_time_dfs)
````

2. BFS sử dụng nhiều bộ nhớ hơn DFS do phải lưu toàn bộ các đỉnh kề trong hàng đợi ở mỗi bước mở rộng.
