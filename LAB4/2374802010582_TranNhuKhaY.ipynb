## L thuyt
Thut ton di truyn (Genetic Algorithm - GA) l mt phng php tm kim da trn c ch tin ha t nhin, ly cm hng t thuyt tin ha ca Darwin. GA c s dng  tm li gii gn ti u cho cc bi ton ti u ha phc tp.

### Cc bc chnh ca thut ton:
1. **Khi to qun th**: To ngu nhin mt tp hp cc c th (gii php tim nng).
2. **nh gi  thch nghi (Fitness Evaluation)**: Tnh gi tr hm mc tiu cho mi c th.
3. **La chn (Selection)**: Chn cc c th tt nht  lai ghp (thng s dng cc phng php nh Tournament Selection hoc Roulette Wheel).
4. **Lai ghp (Crossover)**: Kt hp hai c th  to ra c th mi.
5. **t bin (Mutation)**: Thay i ngu nhin mt s gen  duy tr tnh a dng.
6. **Lp li**: Quay li bc 2 cho n khi t iu kin dng (s th h hoc gi tr fitness mong mun).



### ng dng
- Ti u ha hm s.
- Bi ton lp lch.
- Thit k k thut, machine learning, v.v.

## Cc v d minh ha

2. Tm gi tr ti thiu ca hm s $g(x, y) = x^2 + y^2$ trong khong $x, y \in [-5, 5]$.

Ngoi ra, chng ta s trc quan ha s thay i ca gi tr fitness qua cc th h.


## V d 1: Ti u ha hm mt bin

### Bi ton

$$  
t gi tr ln nht trong khong  
$$
x \in [0, 10].
$$

# M gi
```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection:
        - Chn ngu nhin tournament_size c th trong qun th
        - Chn c th c fitness cao nht lm cha m 1
    Chn cha m 2 bng Tournament Selection tng t
    
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi c gi tr trung bnh ca cha m
        Ngc li:
            Con mi l cha m 1
            
    Thc hin t bin cho con mi vi xc sut mutation_rate:
        Nu t bin:
            Gn con mi gi tr ngu nhin trong [min_val, max_val]
            
    Thm con mi vo qun th mi
    
Cp nht qun th hin ti = qun th mi
In ra c th tt nht v gi tr fitness tng ng ca th h


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return -(x**2) + 10*x + 50  # cc phng trnh

# Khi to qun th
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# La chn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx] 
# Lai ghp
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# t bin
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thut ton di truyn
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
        
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)                                  
            new_population.append(child)              
                                                                
        population = new_population           
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Th h {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr f(x) tt nht')
    plt.title('S thay i ca f(x) qua cc th h')
    plt.show()
    
    return best_individual, best_fitness

# Chy thut ton
best_x, best_f = genetic_algorithm_example1()
print(f'\nKt qu cui cng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

## V d 2: Ti u ha hm hai bin

### Bi ton  
Tm $(x, y)$ sao cho  
$$
g(x, y) = x^2 + y^2
$$  
t gi tr nh nht trong khong  
$$
x, y \in [-5, 5].
$$


### M gi 

```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection trn qun th hin ti
    Chn cha m 2 bng Tournament Selection trn qun th hin ti
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi bng trung bnh gi tr (x, y) ca cha m
        Ngc li:
            Con mi l cha m 1
    Thc hin t bin cho con mi vi xc sut mutation_rate tng bin:
        Nu t bin:
            Gn gi tr ngu nhin trong [min_val, max_val] cho bin tng ng
    Thm con mi vo qun th mi

Cp nht qun th hin ti = qun th mi
In kt qu tt nht th h

````python
# Hm mc tiu
def fitness_function_2d(x, y):
    return -(x**2 + y**2)  # S dng du m v GA tm max

# Khi to qun th
def initialize_population_2d(pop_size, min_val, max_val):
    return [(random.uniform(min_val, max_val), random.uniform(min_val, max_val)) for _ in range(pop_size)]

# La chn
def select_parents_2d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover_2d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return ((parent1[0] + parent2[0]) / 2, (parent1[1] + parent2[1]) / 2)
    return parent1

# t bin
def mutate_2d(individual, mutation_rate=0.1, min_val=-5, max_val=5):
    x, y = individual
    if random.random() < mutation_rate:
        x = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        y = random.uniform(min_val, max_val)
    return (x, y)

# Thut ton di truyn
def genetic_algorithm_example2(pop_size=50, generations=100, min_val=-5, max_val=5):
    population = initialize_population_2d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_2d(x, y) for x, y in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(-best_fitness)  # Chuyn v gi tr thc
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_2d(population, fitness_values)
            parent2 = select_parents_2d(population, fitness_values)
            child = crossover_2d(parent1, parent2)
            child = mutate_2d(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
        print(f'Th h {generation + 1}: (x, y) = ({best_individual[0]:.4f}, {best_individual[1]:.4f}), g(x, y) = {-best_fitness:.4f}')
    
    best_fitness = max([fitness_function_2d(x, y) for x, y in population])
    best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr g(x, y) tt nht')
    plt.title('S thay i ca g(x, y) qua cc th h')
    plt.show()
    
    return best_individual, -best_fitness

# Chy thut ton
best_xy, best_g = genetic_algorithm_example2()
print(f'\nKt qu cui cng: (x, y) = ({best_xy[0]:.4f}, {best_xy[1]:.4f}), g(x, y) = {best_g:.4f}')
````

# Bi tp 1
# Thay i hm mc tiu**: Th p dng thut ton cho hm $h(x) = sin(x) + cos(x)$ trong khong $x \in [0, 2\pi]$. So snh kt qu vi nghim chnh xc.

## Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

## Mc tiu
- Ti u hm:
  $$
  h(x) = \sin(x) + \cos(x)
  $$
  trong min:
  $$
  x \in [0, 2\pi]
  $$
- Tm gi tr **x** sao cho **h(x)** t gi tr ln nht.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##  tng thut ton di truyn (GA)

1. **M ha**: Biu din gi tr \( x \) bng chui nh phn di 16 bit.  
2. **Gii m**: Chuyn t chui nh phn sang gi tr thc trong khong \([0, 2\pi]\).  
3. **Fitness**: S dng chnh gi tr \( h(x) \) lm  thch nghi.  
4. **Chn lc**: Roulette Wheel Selection.  
5. **Lai ghp**: Lai 1 im vi xc sut cao.  
6. **t bin**: Lt bit vi xc sut thp.

---


````python
import numpy as np
import random
import math
import matplotlib.pyplot as plt

num_bits = 16
min_val = 0
max_val = 2 * math.pi

# Hm fitness
def fitness_function(x):
    return math.sin(x) + math.cos(x)

# M ha s thc thnh chui nh phn
def encode(x):
    max_int = 2**num_bits - 1
    int_val = int((x - min_val) / (max_val - min_val) * max_int)
    return format(int_val, f'0{num_bits}b')

# Gii m chui nh phn thnh s thc
def decode(binary_str):
    max_int = 2**num_bits - 1
    int_val = int(binary_str, 2)
    x = min_val + (int_val / max_int) * (max_val - min_val)
    return x

# Khi to qun th nh phn ngu nhin
def initialize_population(pop_size):
    return [format(random.randint(0, 2**num_bits - 1), f'0{min_val}b') for _ in range(pop_size)]

# Tnh fitness cho qun th
def evaluate_population(population):
    return [fitness_function(decode(ind)) for ind in population]

# Chn lc bng Roulette Wheel
def roulette_wheel_selection(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness == 0:
        return random.choice(population)
    probabilities = [f / total_fitness for f in fitness_values]
    return random.choices(population, weights=probabilities, k=1)[0]

# Lai ghp 1 im
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        point = random.randint(1, num_bits - 1)
        child = parent1[:point] + parent2[point:]
        return child
    return parent1

# t bin: lt bit
def mutate(binary_str, mutation_rate=0.01):
    mutated = ''
    for bit in binary_str:
        if random.random() < mutation_rate:
            mutated += '0' if bit == '1' else '1'
        else:
            mutated += bit
    return mutated

# Thut ton di truyn
def genetic_algorithm(pop_size=50, generations=100):
    population = initialize_population(pop_size)
    best_fitness_history = []

    for gen in range(generations):
        fitness_values = evaluate_population(population)
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)

        new_population = []
        for _ in range(pop_size):
            p1 = roulette_wheel_selection(population, fitness_values)
            p2 = roulette_wheel_selection(population, fitness_values)
            child = crossover(p1, p2)
            child = mutate(child)
            new_population.append(child)

        population = new_population

        best_idx = np.argmax(fitness_values)
        best_x = decode(population[best_idx])
        print(f'Th h {gen+1}: x = {best_x:.4f}, h(x) = {best_fitness:.4f}')

    final_fitness = evaluate_population(population)
    best_idx = np.argmax(final_fitness)
    best_x = decode(population[best_idx])
    best_fx = final_fitness[best_idx]

    # V biu 
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel("Th h")
    plt.ylabel("Fitness tt nht")
    plt.title("Tin ha ca h(x)")
    plt.grid(True)
    plt.show()

    return best_x, best_fx

# Chy thut ton
best_x, best_f = genetic_algorithm()
print(f"\nKt qu cui cng: x = {best_x:.4f}, h(x) = {best_f:.4f}")
````

# Bi tp 2: Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

---

## Mc tiu

Ti u hm:
$$
h(x) = \sin(x) + \cos(x)
$$
trong khong:
$$
x \in [0, 2\pi]
$$

> Tm gi tr $x$ sao cho $h(x)$ t **gi tr ln nht**.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##   tng thut ton di truyn (GA)

| Thnh phn        | Chi tit                                                                 |
|------------------|-------------------------------------------------------------------------|
| **M ha**        | Dng chui nh phn 16 bit  biu din gi tr $x \in [0, 2\pi]$         |
| **Gii m**       | nh x nh phn sang gi tr thc: $x = 0 + \dfrac{\text{decimal}}{2^{16}-1} \cdot 2\pi$ |
| **Fitness**       | Chnh l $h(x) = \sin(x) + \cos(x)$                                     |
| **Chn lc**      | Roulette Wheel Selection                                                |
| **Lai ghp**      | 1 im ct vi xc sut 0.8                                              |
| **t bin**      | Xc sut 5% (0.05)                                                       |

---

````python
import numpy as np
import random
import math
import matplotlib.pyplot as plt

# 1) Tham s GA
num_bits = 16
min_val  = 0
max_val  = 2 * math.pi
mutation_rate = 0.05      # 5%
crossover_rate = 0.8      # 80%

# 2) Hm fitness
def fitness_function(x):
    return math.sin(x) + math.cos(x)

# 3) M ha / Gii m
def encode(x):
    max_int = 2**num_bits - 1
    int_val = int((x - min_val) / (max_val - min_val) * max_int)
    return format(int_val, f'0{num_bits}b')

def decode(bitstr):
    max_int = 2**num_bits - 1
    int_val = int(bitstr, 2)
    return min_val + (int_val / max_int) * (max_val - min_val)

# 4) Khi to qun th
def initialize_population(pop_size):
    return [
        format(random.randint(0, 2**num_bits - 1), f'0{num_bits}b')
        for _ in range(pop_size)
    ]

# 5) nh gi, chn lc, lai ghp, t bin
def evaluate_population(pop):
    return [fitness_function(decode(ind)) for ind in pop]

def roulette_wheel(pop, fitness_vals):
    total_fit = sum(fitness_vals)
    if total_fit == 0:
        return random.choice(pop)
    probs = [f / total_fit for f in fitness_vals]
    return random.choices(pop, weights=probs, k=1)[0]

def crossover(p1, p2, rate=crossover_rate):
    if random.random() < rate:
        cut = random.randint(1, num_bits - 1)
        return p1[:cut] + p2[cut:]
    return p1

def mutate(bitstr, rate=mutation_rate):
    return ''.join(
        ('0' if b == '1' else '1') if random.random() < rate else b
        for b in bitstr
    )

# 6) Thut ton di truyn
def genetic_algorithm(pop_size=50, generations=100):
    population = initialize_population(pop_size)
    best_hist  = []

    for gen in range(generations):
        fitness_vals = evaluate_population(population)

        # lu c th tt nht hin ti
        best_idx   = np.argmax(fitness_vals)
        best_x     = decode(population[best_idx])
        best_fit   = fitness_vals[best_idx]
        best_hist.append(best_fit)
        print(f'Th h {gen+1:3d}: x = {best_x:.5f}, h(x) = {best_fit:.5f}')

        # to qun th mi
        new_pop = []
        for _ in range(pop_size):
            p1 = roulette_wheel(population, fitness_vals)
            p2 = roulette_wheel(population, fitness_vals)
            child = crossover(p1, p2)
            child = mutate(child)
            new_pop.append(child)

        population = new_pop

    # kt qu cui
    final_fitness = evaluate_population(population)
    best_idx   = np.argmax(final_fitness)
    best_x     = decode(population[best_idx])
    best_hx    = final_fitness[best_idx]

    # biu 
    plt.plot(range(1, generations+1), best_hist)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr h(x) tt nht')
    plt.title('Qu trnh hi t GA')
    plt.grid(True)
    plt.show()

    return best_x, best_hx

# 7) Chy th
best_x, best_hx = genetic_algorithm()
print(f'\n Kt qu cui cng: x  {best_x:.5f}, h(x)  {best_hx:.5f}')

````

# Bi Tp 3

# Thay i tham s thut ton di truyn v phn tch tc  hi t

**Mc tiu:** Th thay i cc tham s chnh ca thut ton di truyn gm:

- Kch thc qun th: `pop_size`
- S th h: `generations`
- T l t bin: `mutation_rate`
- T l lai ghp: `crossover_rate`

v phn tch nh hng ca tng tham s n tc  hi t ca thut ton.

---

## Cc tham s v vai tr

| Tham s         |  ngha                                  | nh hng chnh                                                                                         |
|-----------------|-----------------------------------------|--------------------------------------------------------------------------------------------------------|
| `pop_size`      | S c th trong qun th                 | Qun th ln gip khm ph khng gian tm kim tt hn nhng tn nhiu thi gian tnh ton. Qun th nh hi t nhanh nhng d b kt cc tr cc b. |
| `generations`   | S vng lp ca thut ton               | S th h cng nhiu th thut ton c nhiu c hi ci thin nghim, tuy nhin thi gian chy cng tng. |
| `mutation_rate` | Xc sut t bin mi bit                | t bin cao gip tng a dng gene, trnh b kt trong cc tr cc b, nhng nu qu cao lm qu trnh tm kim nhiu lon. |
| `crossover_rate`| Xc sut lai ghp gia hai c th       | Lai ghp gip phi hp gen tt, tng kh nng tm nghim tt. Nu qu thp, qun th km a dng; qu cao c th ph hy cu trc tt ca c th. |

---

## Phn tch nh hng

- **`pop_size`**: Kch thc qun th ln gip tm kim ton cc tt hn nhng chm hn, kch thc nh lm thut ton hi t nhanh nhng d mc kt.
- **`generations`**: S th h ln tng kh nng ci tin kt qu nhng lm tng thi gian tnh ton.
- **`mutation_rate`**: T l t bin va phi gip duy tr a dng di truyn, trnh kt cc b; qu cao hoc qu thp u lm gim hiu qu.
- **`crossover_rate`**: T l lai ghp hp l gip khai thc tt thng tin di truyn, tng hiu qu hi t.

---




````python
import numpy as np
import math
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return math.sin(x) + math.cos(x)

# Khi to qun th ngu nhin
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# La chn (Roulette Wheel Selection)
def select_parents(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness == 0:
        return random.choice(population)
    probabilities = [f / total_fitness for f in fitness_values]
    return random.choices(population, weights=probabilities, k=1)[0]

# Lai ghp trung bnh ti 1 im
def crossover(parent1, parent2, crossover_rate):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# t bin (ly ngu nhin trong on gi tr)
def mutate(individual, mutation_rate, min_val, max_val):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thut ton di truyn
def genetic_algorithm(pop_size=30, generations=60,
                      mutation_rate=0.12, crossover_rate=0.75,
                      min_val=0, max_val=2 * math.pi,
                      verbose=True, show_plot=True):

    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []

    for gen in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)

        # In thng tin th h
        if verbose:
            best_x = population[np.argmax(fitness_values)]
            print(f'Th h {gen + 1:3d}: x = {best_x:.5f}, h(x) = {best_fitness:.5f}')

        # To th h mi
        new_population = []
        for _ in range(pop_size):
            p1 = select_parents(population, fitness_values)
            p2 = select_parents(population, fitness_values)
            child = crossover(p1, p2, crossover_rate)
            child = mutate(child, mutation_rate, min_val, max_val)
            new_population.append(child)
        population = new_population

    # Tng kt
    final_fitness = [fitness_function(x) for x in population]
    best_idx = np.argmax(final_fitness)
    best_x = population[best_idx]
    best_fx = final_fitness[best_idx]


    plt.plot(range(1, generations + 1), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr f(x) tt nht')
    plt.title('Tin ha gi tr hm f(x)')
    plt.grid(True)
    plt.show()

    return best_x, best_fx

# 7) Chy th
best_x, best_hx = genetic_algorithm()
print(f'\n Kt qu cui cng: x  {best_x:.5f}, h(x)  {best_hx:.5f}')
````

---

## Mc tiu

##  Cc phng php la chn ph bin

| Phng php                | M t                                                                                   | u im                              | Nhc im                          |

- Tuy nhin c th lm gim a dng gen nu c th u t chim u th qu ln, d ri vo cc tr cc b.
- Phng php **Tournament** kim sot tt s a dng nhng c th chm hn trong mt s trng hp.
- Phng php **Random** n gin nhng t hiu qu trong ti u nng cao.

````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return -(x**2) + 10*x + 50  # nh ti x = 5, f(x) = 75

# Khi to qun th
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# Tournament Selection
def tournament_selection(population, fitness_values, k=3):
    selected = random.sample(range(len(population)), k)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Roulette Wheel Selection
def roulette_selection(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness == 0:
        return random.choice(population)
    probabilities = [f / total_fitness for f in fitness_values]
    return random.choices(population, weights=probabilities, k=1)[0]

# Random Selection
def random_selection(population, *_):
    return random.choice(population)

# Lai ghp trung bnh
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# t bin
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thut ton di truyn chnh
def genetic_algorithm(pop_size=50, generations=100,
                      min_val=0, max_val=10,
                      selection_method='tournament'):
    
    # Chn hm chn lc
    if selection_method == 'tournament':
        select = lambda pop, fit: tournament_selection(pop, fit)
    elif selection_method == 'roulette':
        select = lambda pop, fit: roulette_selection(pop, fit)
    elif selection_method == 'random':
        select = lambda pop, fit: random_selection(pop)
    else:
        raise ValueError("Phng php chn lc khng hp l")

    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []

    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)

        new_population = []
        for _ in range(pop_size):
            parent1 = select(population, fitness_values)
            parent2 = select(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child, min_val=min_val, max_val=max_val)
            new_population.append(child)
        population = new_population

    best_idx = np.argmax([fitness_function(x) for x in population])
    best_x = population[best_idx]
    best_fx = fitness_function(best_x)

    return best_x, best_fx, best_fitness_history

methods = ['tournament', 'roulette', 'random']
colors = ['blue', 'green', 'red']
histories = []

for method in methods:
    print(f"\nphng php: {method}")
    best_x, best_fx, history = genetic_algorithm(selection_method=method)
    print(f"Kt qu: x  {best_x:.4f}, f(x)  {best_fx:.4f}")
    histories.append((method, history))

    # V biu  ring cho tng phng php
    plt.figure(figsize=(8, 5))
    plt.plot(range(1, len(history) + 1), history, color=colors[methods.index(method)])
    plt.title(f" {method.capitalize()} Selection")
    plt.xlabel("Th h")
    plt.ylabel("Fitness tt nht")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
````

# Trc quan ha qun th  V biu  phn tn (scatter plot) th hin s phn b ca qun th qua cc th h

---

##  Mc tiu

- Trc quan ha s phn b ca qun th c th trong qu trnh tin ha ca thut ton di truyn.
- Dng biu  phn tn (scatter plot)  th hin v tr cc c th trn khng gian bin s qua tng th h.
- Gip quan st r tin trnh hi t, a dng di truyn, v s thay i qun th theo thi gian.

---

##  tng thc hin

- Vi bi ton ti u hm $k(x, y, z) = x^2 + y^2 + z^2$, khng gian tm kim l 3 chiu.
- Mi c th l mt im trong khng gian $(x, y, z)$.
-  mi th h, ly v tr gii m ca ton b c th, v scatter plot.
- Nu khng th trc quan 3D trc tip, c th v scatter 2D tng cp bin: $(x, y)$, $(y, z)$, $(x, z)$ qua cc th h.

---

````python
import numpy as np
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Hm fitness cn ti u (ti thiu ha)
def fitness_function(x, y, z):
    return -(x**2 + y**2 + z**2)  # t du m  GA tm max

# Khi to qun th 3D
def initialize_population(pop_size, min_val, max_val):
    return [(
        random.uniform(min_val, max_val),
        random.uniform(min_val, max_val),
        random.uniform(min_val, max_val)
    ) for _ in range(pop_size)]

# Chn lc ngu nhin
def select_random(population):
    return random.choice(population)

# Lai ghp
def crossover(p1, p2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return tuple((a + b) / 2 for a, b in zip(p1, p2))
    return p1

# t bin
def mutate(ind, mutation_rate=0.1, min_val=-10, max_val=10):
    return tuple(
        random.uniform(min_val, max_val) if random.random() < mutation_rate else gene
        for gene in ind
    )

# Thut ton di truyn + ghi li qun th theo th h
def genetic_algorithm_3d(pop_size=100, generations=50, min_val=-10, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    history = []

    for gen in range(generations):
        fitness_values = [fitness_function(x, y, z) for x, y, z in population]
        best_idx = np.argmax(fitness_values)
        best = population[best_idx]
        print(f"Th h {gen+1:02d}: x={best[0]:.2f}, y={best[1]:.2f}, z={best[2]:.2f}, f={-fitness_values[best_idx]:.4f}")
        
        history.append(population.copy())  # lu li qun th

        new_population = []
        for _ in range(pop_size):
            p1 = select_random(population)
            p2 = select_random(population)
            child = crossover(p1, p2)
            child = mutate(child, min_val=min_val, max_val=max_val)
            new_population.append(child)

        population = new_population

    return history

# V scatter 3D cho tng th h
def visualize_evolution_3d(history, skip=10):
    for gen, pop in enumerate(history):
        if gen % skip != 0:
            continue  # gim s lng hnh v nu qu nhiu
        xs = [x for x, y, z in pop]
        ys = [y for x, y, z in pop]
        zs = [z for x, y, z in pop]

        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.scatter(xs, ys, zs, c='blue', alpha=0.6)
        ax.set_title(f'Phn b qun th  Th h {gen+1}')
        ax.set_xlim([-10, 10])
        ax.set_ylim([-10, 10])
        ax.set_zlim([-10, 10])
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_zlabel('z')
        plt.show()

# Chy thut ton & trc quan ha
history = genetic_algorithm_3d()
visualize_evolution_3d(history, skip=10)
````
